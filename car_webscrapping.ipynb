```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
url  = 'https://www.zigwheels.com/newcars/best-cars-in-india'

```

```python
def doc(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.text,'html.parser')
  return soup
doc1=doc(url)
```

```python
Cars_dict = {}
cars_no = 0
```

```python
cars =doc1.find_all("div", {"class": "p-10 mke-ryt rel"})
```

```python
for car in cars:
          url = 'https://www.zigwheels.com/newcars/best-cars-in-india'
          page = requests.get(url)
          soup = BeautifulSoup(page.text, 'html.parser')

          name      = car.find('a', {'class':"fnt-14 lnk-hvr block of-hid ht-name"}).text
          price = car.find('div', {'class': 'b fnt-black fnt-14'}).text
          details = car.find('div', {'class': 'clr-try fnt-12 pb-10 ht-spec of-hid'}).text

          cars_no = cars_no + 1
          Cars_dict[cars_no] = [name, price,details]
cars_dict_df = pd.DataFrame.from_dict(Cars_dict, orient='index',columns=['name','price','details'])

cars_dict_df     
```

```python
url=[]
for car in doc1.find_all('a', class_="fnt-14 lnk-hvr block of-hid ht-name"):
    url.append(car['href'])
```

```python
for car in url:
    page2=doc(car)
    fuel=page2.find('span',itemprop="fuelType").text
   
    engine=page2.find('span',itemprop="name").text
    # print(fuel,engine)

    
```

```python
cars_dict_df['fuel'] =fuel
cars_dict_df['engine'] =engine
```
